```python
import pandas as pd

df = pd.read_excel('/content/netflix engagement report.xlsx', sheet_name='main_formated 2023')
```

```python
df.head(10)
```

```python
import regex as re

def is_japanese(text):
  pattern = re.compile(r'([\p{IsHan}\p{IsBopo}\p{IsHira}\p{IsKatakana}]+)', re.UNICODE)
  input = f'u{text}'
  result = pattern.findall(input)
  # print(result)
  if result != []:
    return 1
  return 0
```

```python
is_japanese('Black Clover: Sword of the Wizard King')
```

```python
df['japanese'] = df['Title'].apply(is_japanese)
df.head()
```

```python
japenese_cleaned = df[df['japanese']==1]
```

```python
japenese_cleaned.to_csv("japenese_cleaned.csv",index=False)
```
